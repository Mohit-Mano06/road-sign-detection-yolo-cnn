```python
import os
from pathlib import Path

train_path = Path(r'C:\Users\delta\OneDrive\Desktop\Minor_Project_Code\yolo_roboflow\Road-Sign-detection-10\train')

def check_normalization(file_path):
    with open(file_path, 'r') as f:
        for line in f:
            values = list(map(float, line.strip().split()[1:]))  # Skip class ID
            if any(v < 0 or v > 1 for v in values):
                return False
    return True

# Check first 100 label files
for label_file in list(train_path.glob('*.txt'))[:100]:
    if not check_normalization(label_file):
        print(f"File {label_file.name} contains non-normalized values")

print("Normalization check complete")
```

```python
from PIL import Image
import os
from pathlib import Path

train_path = Path(r'C:\Users\delta\OneDrive\Desktop\Minor_Project_Code\yolo_roboflow\Road-Sign-detection-10\train')

def check_image(file_path):
    try:
        with Image.open(file_path) as img:
            img.verify()
        return True
    except:
        return False

corrupted_images = []
for img_file in train_path.glob('*.jpg'):
    if not check_image(img_file):
        corrupted_images.append(img_file.name)

print(f"Number of corrupted images: {len(corrupted_images)}")
if corrupted_images:
    print("Sample of corrupted images:", corrupted_images[:5])
```

```python
import numpy as np

def analyze_labels(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()
    
    if not lines:
        return "empty"
    
    values = np.array([list(map(float, line.strip().split()[1:])) for line in lines])
    
    if np.any(values < 0) or np.any(values > 1):
        return "out_of_range"
    
    if np.any(values[:, 2:] > 1):  # width and height should not be > 1
        return "invalid_size"
    
    return "ok"

label_status = {"empty": 0, "out_of_range": 0, "invalid_size": 0, "ok": 0}

for label_file in train_path.glob('*.txt'):
    status = analyze_labels(label_file)
    label_status[status] += 1

print("Label analysis results:")
for status, count in label_status.items():
    print(f"{status}: {count}")
```

```python
from ultralytics import YOLO

model = YOLO('yolov8s.pt')  # Using nano model for quicker training

org_data_yaml_path = r'C:\Users\delta\OneDrive\Desktop\Minor_Project_Code\yolo_roboflow\Road-Sign-detection-10\data.yaml'

results = model.train(
    data=str(org_data_yaml_path),
    epochs=50,
    imgsz=640,
    batch=2,
    device='0',
    lr0=1e-4,
    lrf=1e-5,
    verbose=True,
    amp=True,
    warmup_epochs=5,
    mosaic=0.0,
    mixup=0.0,
)

val_results = model.val()
print(f"Validation results: {val_results}")
```
